# Adaboost (Boosting) script on the SimpleNN XGBoost and SVM models
This script is aimed at building and evaluating several machine learning models for the prediction of a binary outcome variable (DEATH_EVENT) using a dataset on heart failure clinical records. The script first imports necessary libraries such as pandas, numpy, torch, and various modules from scikit-learn and XGBoost for model building, evaluation, and visualization.

Here's a breakdown of the main steps in the script:

Data Loading and Preprocessing:

The dataset is loaded using pandas from a CSV file.
Some variables are renamed for clarity.
Standardization is applied to selected columns using StandardScaler from scikit-learn.
The data is split into training, validation, and test sets.
SMOTE (Synthetic Minority Over-sampling Technique) is applied to balance the class distribution in the training set.
Model Building:

Three types of models are trained: a simple neural network (SimpleNN), XGBoost, and Support Vector Machine (SVM).
For SimpleNN, an MLPClassifier is used.
For XGBoost, a parameter grid is defined for hyperparameter tuning using GridSearchCV, and an XGBClassifier is utilized.
For SVM, a parameter grid is defined similarly, and a grid search is performed to find the best parameters.
AdaBoost is employed to enhance the performance of each base model. AdaBoost iteratively improves the model by focusing on misclassified instances.
Model Evaluation:

Each model (SimpleNN, XGBoost, SVM) boosted with AdaBoost is evaluated on the validation set using metrics such as accuracy, precision, and recall.
Similarly, each model is evaluated on the test set to assess its generalization performance.
Confusion matrices are generated for each boosted model on the test set, visualizing the classification results.
Comparison:

The script prints out the evaluation metrics for each boosted model (validation and test accuracy, precision, and recall).
Additionally, it creates confusion matrices to visually compare the performance of the boosted models on the test set.
Boosting methods like AdaBoost are employed to enhance the performance of weak base learners iteratively. In this script, AdaBoost is applied to each base model to improve its predictive performance. By sequentially adjusting the weights of misclassified instances, AdaBoost focuses on areas where the base model performs poorly, potentially leading to better overall performance.

Comparing boosted models to their original counterparts, you would expect to see improvements in performance metrics such as accuracy, precision, and recall, as boosting aims to reduce bias and variance, leading to better generalization on unseen data. Additionally, the confusion matrices allow for a visual comparison of the classification results, helping to assess the effectiveness of boosting in improving model predictions.

# Adaboost

In the script provided, AdaBoost is like having a team of specialists working together to solve a problem. Each specialist, called a base learner, has their own unique way of understanding the data. However, they may not be perfect and might make mistakes. AdaBoost's job is to coordinate these specialists in such a way that they complement each other's strengths and weaknesses. It does this by paying more attention to the areas where the specialists struggle, kind of like giving them extra practice in those specific areas. As a result, AdaBoost helps the specialists improve over time, leading to a more accurate overall solution.

In practical terms, AdaBoost starts by training a simple model, like a decision tree, on the data. It then evaluates how well this model performs and identifies which data points it struggles with. AdaBoost then trains another model, placing extra emphasis on these difficult data points, so the new model focuses more on getting them right. This process repeats multiple times, with each new model learning from the mistakes of the previous ones. Eventually, AdaBoost combines all these models together into one powerful solution that is better at making predictions than any individual model on its own.

In the script, AdaBoost is used to improve the performance of different machine learning models, like neural networks and support vector machines, by guiding them to focus on the most challenging aspects of the data. By doing so, AdaBoost helps these models become more accurate and reliable, ultimately leading to better predictions for tasks like predicting heart failure outcomes. The results of AdaBoost are then compared to the original models to see how much they've improved. This way, AdaBoost acts as a supportive mechanism that enhances the capabilities of these models, making them more effective in real-world applications.

```python
import pandas as pd
import numpy as np
import torch
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier  
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from imblearn.over_sampling import SMOTE
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('../heart_failure_clinical_records_dataset.csv')

# Rename variables for clarity
target = 'DEATH_EVENT'
features = df.columns[df.columns != target]

# List of columns to be standardized
columns_to_standardize = df.columns.difference(['anemia', 'diabetes', 'high_blood_pressure', 'sex', 'smoking', 'DEATH_EVENT'])

# Instantiate StandardScaler
scaler = StandardScaler()

# Fit and transform the selected columns
df[columns_to_standardize] = scaler.fit_transform(df[columns_to_standardize])

# Split the data into training, validation, and test sets
df_train, df_temp = train_test_split(df, test_size=0.4, random_state=42)
df_val, df_test = train_test_split(df_temp, test_size=0.1, random_state=42)

# Extract features and target variable
X_train = df_train.drop(target, axis=1).values
y_train = df_train[target].values

# Apply SMOTE to balance the class distribution
smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)

X_val = df_val.drop(target, axis=1).values
y_val = df_val[target].values

X_test = df_test.drop(target, axis=1).values
y_test = df_test[target].values

# Convert data to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32)

X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val, dtype=torch.float32)

X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32)

# Simple Neural Network (SimpleNN)
simpleNN = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)

# Decision Tree as a base estimator for AdaBoost
base_estimator = DecisionTreeClassifier(max_depth=1, random_state=42)

# AdaBoost for Simple Neural Network
adaboost_simpleNN = AdaBoostClassifier(base_estimator=base_estimator, n_estimators=50, random_state=42, algorithm='SAMME')
adaboost_simpleNN.fit(X_train, y_train)

# Evaluate on validation set
val_predictions_simpleNN_adaboost = adaboost_simpleNN.predict(X_val)
val_accuracy_simpleNN_adaboost = accuracy_score(y_val, val_predictions_simpleNN_adaboost)
val_precision_simpleNN_adaboost = precision_score(y_val, val_predictions_simpleNN_adaboost)
val_recall_simpleNN_adaboost = recall_score(y_val, val_predictions_simpleNN_adaboost)

print(f'AdaBoost SimpleNN Validation Accuracy: {val_accuracy_simpleNN_adaboost:.4f}')
print(f'AdaBoost SimpleNN Validation Precision: {val_precision_simpleNN_adaboost:.4f}')
print(f'AdaBoost SimpleNN Validation Recall: {val_recall_simpleNN_adaboost:.4f}')

# Evaluate on test set
test_predictions_simpleNN_adaboost = adaboost_simpleNN.predict(X_test)
test_accuracy_simpleNN_adaboost = accuracy_score(y_test, test_predictions_simpleNN_adaboost)
test_precision_simpleNN_adaboost = precision_score(y_test, test_predictions_simpleNN_adaboost)
test_recall_simpleNN_adaboost = recall_score(y_test, test_predictions_simpleNN_adaboost)

print(f'AdaBoost SimpleNN Test Accuracy: {test_accuracy_simpleNN_adaboost:.4f}')
print(f'AdaBoost SimpleNN Test Precision: {test_precision_simpleNN_adaboost:.4f}')
print(f'AdaBoost SimpleNN Test Recall: {test_recall_simpleNN_adaboost:.4f}')

# Create confusion matrix for AdaBoost SimpleNN on the test set
conf_matrix_simpleNN_adaboost = confusion_matrix(y_test, test_predictions_simpleNN_adaboost)
total_samples = len(y_test)
sns.heatmap(conf_matrix_simpleNN_adaboost / total_samples * 100, annot=True, fmt='.2f', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Percentage) - AdaBoost SimpleNN')
plt.show()

# XGBoost
param_grid_xgb = {
    'learning_rate': [0.90, 0.99, 0.9999, 0.999999],
    'max_depth': [3, 4, 5, 6],
    'n_estimators': [114, 116],
    'subsample': [0.8, 0.9, 1.0],
}

xgb_classifier = XGBClassifier()
adaboost_xgb = AdaBoostClassifier(base_estimator=xgb_classifier, n_estimators=50, random_state=42)
adaboost_xgb.fit(X_train, y_train)

# Evaluate on validation set
val_predictions_xgb_adaboost = adaboost_xgb.predict(X_val)
val_accuracy_xgb_adaboost = accuracy_score(y_val, val_predictions_xgb_adaboost)
val_precision_xgb_adaboost = precision_score(y_val, val_predictions_xgb_adaboost)
val_recall_xgb_adaboost = recall_score(y_val, val_predictions_xgb_adaboost)

print(f'AdaBoost XGBoost Validation Accuracy: {val_accuracy_xgb_adaboost:.4f}')
print(f'AdaBoost XGBoost Validation Precision: {val_precision_xgb_adaboost:.4f}')
print(f'AdaBoost XGBoost Validation Recall: {val_recall_xgb_adaboost:.4f}')

# Evaluate on test set
test_predictions_xgb_adaboost = adaboost_xgb.predict(X_test)
test_accuracy_xgb_adaboost = accuracy_score(y_test, test_predictions_xgb_adaboost)
test_precision_xgb_adaboost = precision_score(y_test, test_predictions_xgb_adaboost)
test_recall_xgb_adaboost = recall_score(y_test, test_predictions_xgb_adaboost)

print(f'AdaBoost XGBoost Test Accuracy: {test_accuracy_xgb_adaboost:.4f}')
print(f'AdaBoost XGBoost Test Precision: {test_precision_xgb_adaboost:.4f}')
print(f'AdaBoost XGBoost Test Recall: {test_recall_xgb_adaboost:.4f}')

# Create confusion matrix for AdaBoost XGBoost on the test set
conf_matrix_xgb_adaboost = confusion_matrix(y_test, test_predictions_xgb_adaboost)
total_samples = len(y_test)
sns.heatmap(conf_matrix_xgb_adaboost / total_samples * 100, annot=True, fmt='.2f', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Percentage) - AdaBoost XGBoost')
plt.show()

# SVM
param_grid_svm = {'C': [3, 4, 5], 'gamma': [0.1, 0.2, 0.25, 0.001], 'kernel': ['linear', 'poly', 'sigmoid']}
svm_classifier = SVC(random_state=42)

# Perform grid search
grid_search_svm = GridSearchCV(estimator=svm_classifier, param_grid=param_grid_svm, cv=15, scoring='accuracy')
grid_search_svm.fit(X_train, y_train)

# Get best parameters
best_params_svm = grid_search_svm.best_params_

# Create SVM classifier with best parameters
best_svm_classifier = SVC(**best_params_svm, random_state=42)

# AdaBoost for SVM
adaboost_svm = AdaBoostClassifier(base_estimator=best_svm_classifier, n_estimators=50, random_state=42, algorithm='SAMME')
adaboost_svm.fit(X_train, y_train)

# Evaluate on validation set
val_predictions_svm_adaboost = adaboost_svm.predict(X_val)
val_accuracy_svm_adaboost = accuracy_score(y_val, val_predictions_svm_adaboost)
val_precision_svm_adaboost = precision_score(y_val, val_predictions_svm_adaboost)
val_recall_svm_adaboost = recall_score(y_val, val_predictions_svm_adaboost)

print(f'AdaBoost SVM Validation Accuracy: {val_accuracy_svm_adaboost:.4f}')
print(f'AdaBoost SVM Validation Precision: {val_precision_svm_adaboost:.4f}')
print(f'AdaBoost SVM Validation Recall: {val_recall_svm_adaboost:.4f}')

# Evaluate on test set
test_predictions_svm_adaboost = adaboost_svm.predict(X_test)
test_accuracy_svm_adaboost = accuracy_score(y_test, test_predictions_svm_adaboost)
test_precision_svm_adaboost = precision_score(y_test, test_predictions_svm_adaboost)
test_recall_svm_adaboost = recall_score(y_test, test_predictions_svm_adaboost)

print(f'AdaBoost SVM Test Accuracy: {test_accuracy_svm_adaboost:.4f}')
print(f'AdaBoost SVM Test Precision: {test_precision_svm_adaboost:.4f}')
print(f'AdaBoost SVM Test Recall: {test_recall_svm_adaboost:.4f}')

# Create confusion matrix for AdaBoost SVM on the test set
conf_matrix_svm_adaboost = confusion_matrix(y_test, test_predictions_svm_adaboost)
total_samples = len(y_test)
sns.heatmap(conf_matrix_svm_adaboost / total_samples * 100, annot=True, fmt='.2f', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Percentage) - AdaBoost SVM')
plt.show()
```
